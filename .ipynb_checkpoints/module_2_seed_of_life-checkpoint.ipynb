## 2. Constraint Mapping

The Seed of Life introduces the geometry of the **Hexagon** to our framework.

* **Angles:** The centers of the 6 outer circles form a perfect regular hexagon. Each internal angle from the center is exactly $60^\circ$ or $\frac{\pi}{3}$ radians.
* **Tessellation:** The Seed of Life proves that circles, when overlapping by their exact radius $r$, perfectly tile a 2D plane without any distortion. 
* **The Hidden Constant:** By connecting specific intersections within this pattern, we can derive $\sqrt{3}$ (from Module 1) and introduce $\sqrt{2}$ (the diagonal of a square), proving that these constants are inherently linked.

```rust
// Rust Code: SVG Visualization Generator for the Seed of Life
// Using std::f64::consts::PI to calculate the 60-degree radial steps.

fn generate_seed_svg(radius: f64) -> String {
    let width = radius * 5.0;
    let height = radius * 5.0;
    let cx = width / 2.0;
    let cy = height / 2.0;
    
    let mut svg = format!(
        "<svg width='{w}' height='{h}' viewBox='0 0 {w} {h}' xmlns='http://www.w3.org/2000/svg'>\n\
         <rect width='100%' height='100%' fill='#1a1a1a' />\n\
         \n\
         <circle cx='{cx}' cy='{cy}' r='{r}' stroke='#ffffff' stroke-width='2' fill='none' opacity='0.8' />\n",
        w = width, h = height, cx = cx, cy = cy, r = radius
    );

    // 6 Surrounding Circles calculated using Trigonometry (Angles of PI/3)
    use std::f64::consts::PI;
    for i in 0..6 {
        let angle = (i as f64) * (PI / 3.0);
        let x = cx + radius * angle.cos();
        let y = cy + radius * angle.sin();
        
        svg.push_str(&format!(
            "<circle cx='{x}' cy='{y}' r='{r}' stroke='#4a90e2' stroke-width='2' fill='none' opacity='0.6' />\n",
            x = x, y = y, r = radius
        ));
    }

    svg.push_str("</svg>");
    svg
}

let svg_output = generate_seed_svg(100.0);
println!("EVCXR_BEGIN_CONTENT text/html\n{}\nEVCXR_END_CONTENT", svg_output);
```
