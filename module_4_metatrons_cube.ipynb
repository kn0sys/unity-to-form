## 2. Constraint Mapping: The Platonic Solids

Hidden within the 78 intersecting lines of Metatron's Cube are the 2D projections of the five **Platonic Solids**. These are the *only* five shapes in existence where every face, edge, and angle are identical, and where every vertex perfectly touches the inside of a sphere.

* **Tetrahedron:** 4 triangular faces.
* **Hexahedron (Cube):** 6 square faces.
* **Octahedron:** 8 triangular faces.
* **Dodecahedron:** 12 pentagonal faces.
* **Icosahedron:** 20 triangular faces.

**Euler's Formula:** Every Platonic solid obeys the rigid structural limit defined by the equation $V - E + F = 2$ (Vertices minus Edges plus Faces always equals 2). This is the mathematical rule that defines physical closed volume.

```rust
// Rust Code: SVG Visualization Generator for Metatron's Cube
// We calculate the 13 nodes of the Fruit of Life and programmatically
// route the 78 informational connections between them.

use std::f64::consts::PI;

fn generate_metatron_svg(spacing: f64) -> String {
    let width = spacing * 6.0;
    let height = spacing * 6.0;
    let cx = width / 2.0;
    let cy = height / 2.0;
    
    let mut svg = format!(
        "<svg width='{w}' height='{h}' viewBox='0 0 {w} {h}' xmlns='http://www.w3.org/2000/svg'>\n\
         <rect width='100%' height='100%' fill='#1a1a1a' />\n",
        w = width, h = height
    );

    // Array to hold our 13 coordinate centers
    let mut points = Vec::new();
    points.push((cx, cy)); // 1. The Monad (Center)

    // Generate the other 12 points (6 inner, 6 outer)
    for i in 0..6 {
        let angle = (i as f64) * (PI / 3.0);
        // Inner ring
        points.push((cx + spacing * angle.cos(), cy + spacing * angle.sin()));
        // Outer ring
        points.push((cx + 2.0 * spacing * angle.cos(), cy + 2.0 * spacing * angle.sin()));
    }

    // Draw the 78 structural lines of the Matrix
    for i in 0..points.len() {
        for j in (i + 1)..points.len() {
            let (x1, y1) = points[i];
            let (x2, y2) = points[j];
            svg.push_str(&format!(
                "<line x1='{}' y1='{}' x2='{}' y2='{}' stroke='#4a90e2' stroke-width='1.5' opacity='0.3' />\n",
                x1, y1, x2, y2
            ));
        }
    }

    // Draw the 13 spheres of the Fruit of Life
    for (x, y) in points.iter() {
        svg.push_str(&format!(
            "<circle cx='{}' cy='{}' r='{}' stroke='#4a90e2' stroke-width='2.5' fill='#1a1a1a' opacity='0.1' />\n",
            x, y, spacing * 0.5
        ));
    }

    svg.push_str("</svg>");
    svg
}

let svg_output = generate_metatron_svg(80.0);
println!("EVCXR_BEGIN_CONTENT text/html\n{}\nEVCXR_END_CONTENT", svg_output);
```
